```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Design Matrix

cheating_pattern_1 = np.random.choice(3, 100)
```

```python
cheating_pattern_2 = np.random.choice([-1, 1], size=(3, 25))
```

```python
design_matrix = [cheating_pattern_2[i] for i in cheating_pattern_1]
```

```python
plt.matshow(design_matrix)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, n_init=10, init="k-means++")
output_labels = kmeans.fit_predict(design_matrix)
```

```python
plt.scatter(output_labels, cheating_pattern_1)
```
